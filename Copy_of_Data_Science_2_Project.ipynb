# Hypothesis Testing

**1. Cardio burns calories more than HIIT**


*   ***Null Hypothesis:*** The mean calories burned during Cardio sessions is less than or equal HIIT
*   ***Alternative Hypothesis:*** The mean calories burned during Cardio sessions is more than HIIT



```python

cardio_calories = df[df['Workout_Type'] == 'Cardio']['Calories_Burned']
hiit_calories = df[df['Workout_Type'] == 'HIIT']['Calories_Burned']


mean_cardio = cardio_calories.mean()
std_cardio = cardio_calories.std()
n_cardio = len(cardio_calories)

mean_hiit = hiit_calories.mean()
std_hiit = hiit_calories.std()
n_hiit = len(hiit_calories)


pooled_se = np.sqrt((std_cardio**2 / n_cardio) + (std_hiit**2 / n_hiit))

z_score = (mean_cardio - mean_hiit) / pooled_se

p_value = norm.sf(z_score)

print("Z-Score:", z_score)
print("P-Value:", p_value)
```

Based on the Z-Score and P-value provided from the analysis done above, we fail to reject the null hypothesis.
